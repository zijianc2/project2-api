## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

url = "https://en.wikipedia.org/wiki/Oasis_(band)"

response = requests.get(url)

soup = bs4.BeautifulSoup(response.text, "html.parser")

text_content = soup.get_text()

print("Page Text Content (first 500 characters):\n")
print(text_content[:500])
```

```python
#download an image with beautifulsoup and save it in this repository

image = soup.find("img")

if image:
    img_src = image["src"]
    
    if img_src.startswith("//"):
        img_url = "https:" + img_src 
    elif img_src.startswith("/"):
        img_url = "https://en.wikipedia.org" + img_src  
    else:
        img_url = img_src  
    
    img_data = requests.get(img_url).content
    with open("oasis_band.jpg", "wb") as f:
        f.write(img_data)
    
    print("Image downloaded: oasis_band.jpg")
else:
    print("No images found on the page.")
```

```python
#find all the links in a page with beautifulsoup
links = [a["href"] for a in soup.find_all("a", href=True)]

filtered_links = [link for link in links if link.startswith("http")]

print("First 10 links on the page:")
for i, link in enumerate(filtered_links[:10]):
    print(f"{i+1}. {link[:100]}")
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
scripts
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
import requests
from bs4 import BeautifulSoup

scripts["script_text"] = ""

for index, row in scripts.head(10).iterrows():
    url = row["link"]
    
    if pd.notna(url) and url.startswith("http"):
        try:
            response = requests.get(url, timeout=10) 
            soup = BeautifulSoup(response.text, "html.parser")  
            text = soup.get_text()[:1000] 
            
            scripts.at[index, "script_text"] = text
            
            print(f"Downloaded: {row['title']}") 
        except Exception as e:
            print(f"Failed to download {row['title']}: {e}")

scripts[["title", "script_text"]]
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
scripts.to_csv("pudding_texts.csv", index=False)

print("Saved to pudding_texts.csv")
```

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import os
os.makedirs("movie_posters", exist_ok=True)
API_KEY = "028fd577eae6a1a494052d66daa34656"
BASE_URL = "https://api.themoviedb.org/3"
MOVIE_NOW_PLAYING_URL = f"{BASE_URL}/movie/now_playing?api_key={API_KEY}&language=en-US&page=1"
IMAGE_BASE_URL = "https://image.tmdb.org/t/p/w500" 
response = requests.get(MOVIE_NOW_PLAYING_URL)
if response.status_code == 200:
    movies_data = response.json()
    movies = movies_data.get("results", [])[:10]  # 只获取前 10 部电影
else:
    print("Failed to fetch data from TMDB")
    movies = []
movie_list = []
for movie in movies:
    movie_list.append({
        "id": movie["id"],
        "title": movie["title"],
        "release_date": movie["release_date"],
        "overview": movie["overview"],
        "poster_path": movie["poster_path"]
    })
movies_df = pd.DataFrame(movie_list)
movies_df
```

```python
#download the movie posters for 10 of these movies and save them to this repository
for movie in movies:
    poster_path = movie.get("poster_path")
    if poster_path:
        poster_url = f"{IMAGE_BASE_URL}{poster_path}"
        poster_response = requests.get(poster_url)
        
        if poster_response.status_code == 200:
            file_path = f"movie_posters/{movie['title'].replace(' ', '_')}.jpg"
            with open(file_path, "wb") as f:
                f.write(poster_response.content)
            print(f"Downloaded: {movie['title']}")
        else:
            print(f"Failed to download poster for {movie['title']}")
```
